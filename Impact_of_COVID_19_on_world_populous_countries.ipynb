```python
code = raw_data["iso_code"].unique().tolist()
country = raw_data["location"].unique().tolist()
population = raw_data["population"].unique().tolist()
gdp = raw_data["gdp_per_capita"].unique().tolist()
hdi = []
tc = []
td = []
sti = []

for i in country:
    hdi.append(((raw_data.loc[raw_data["location"] == i, "human_development_index"]).sum())/294)
    tc.append((transformed_data.loc[transformed_data["Country"] == i, "Cases"]).sum())
    td.append((transformed_data.loc[transformed_data["Country"] == i, "Deaths"]).sum())
    sti.append((raw_data.loc[raw_data["location"] == i, "stringency_index"]).sum()/294)
```

```python
new_data = pd.DataFrame(list(zip(code,country,population,tc,td,sti,gdp,hdi)),
                           columns = ["Country_code", "Country", "Total_Population", "Total_Cases","Total_deaths","Stringency_Index","GDP", "Human_Index"]
                       )
new_data
```
